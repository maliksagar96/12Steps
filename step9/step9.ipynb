```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def plot2D(x, y, p):


    fig = plt.figure(figsize=(11, 7), dpi=100)
    ax = fig.add_subplot(111, projection='3d')

    X, Y = np.meshgrid(x, y)
    surf = ax.plot_surface(X, Y, p, rstride=1, cstride=1, cmap=cm.viridis,
                           linewidth=0, antialiased=False)

    ax.set_xlim(0, 2)
    ax.set_ylim(0, 1)
    ax.view_init(30, 225)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    plt.show()
```

```python
nx = 100
ny = 100
dx = 2/(nx-1)
dy = 2/(ny-1)

x = np.linspace(0, 2, nx)
y = np.linspace(0, 1, ny)

#Initial Conditions
p = np.zeros([nx,ny])

#Boundary Conditions
p[0, :] = 0
p[:,0] = p[:,1] #Derivative Boundary condition
p[:, -1] = p[:,-2]
p[-1, :] = y
```

```python
dy2 = dy*dy
dx2 = dx*dx
residue = 1
counter = 1
while (residue > 1e-4):
    pn = p.copy()
    for i in range(1,nx-1):
        for j in range(1,ny-1):
            p[i,j] = (dy2*(pn[i+1,j] + pn[i-1,j]) + dx2*(pn[i, j+1] + pn[i, j-1]))/(2*(dx2+dy2))
    p[0, :] = 0
    p[:,0] = p[:,1] #Derivative Boundary condition
    p[:, -1] = p[:,-2]
    p[-1, :] = y
    
    residue = np.abs((np.sum(np.abs(p[:])) - np.sum(np.abs(pn[:])))/(np.sum(np.abs(p[:]))))
    print("Reisdue = ", residue, "counter = ", counter)
    counter+=1
```

```python
plot2D(x,y,p)
```
